```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
direcoes = [0,45,90,135,180]
 
def aplicar_rotacao(ponto,angulo):
    theta = np.radians(angulo)
    matrizRotacao = np.array([[np.cos(theta), -np.sin(theta)],
                               [np.sin(theta), np.cos(theta)]])
    return np.dot(matrizRotacao, ponto)
 
plt.figure(figsize=(10,5))
plt.title("Animação: Personagem olhando em diferentes direções")
for angulo in direcoes:
    posicaoInicial = np.array([2*np.cos(np.radians(angulo)), 2*np.sin(np.radians(angulo))])
    posicaoRotacionada = aplicar_rotacao(posicaoInicial,angulo)
    plt.scatter(posicaoRotacionada[0],posicaoRotacionada[1], label = f'Direção {angulo}', marker = 'o', s=200)
 
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
